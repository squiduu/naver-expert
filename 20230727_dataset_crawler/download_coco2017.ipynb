# Annotation 파일 읽기

```python
import json

with open("/content/annotations/instances_train2017.json", mode="r") as f:
    instances = json.load(f)
```

```python
file_names = sorted(os.listdir("/content/person")) # 이름만 바꿔주세요! ex) drone -> car
for file_name in file_names:
    for image in instances["images"]:
        if image["file_name"] == file_name:
            image_id = image["id"]

            category_ids = set()
            for annotation in instances["annotations"]:
                if annotation["image_id"] == image_id:
                    category_ids.add(annotation["category_id"])

            class_names = []
            for cat_id in category_ids:
                for cat in instances["categories"]:
                    if cat_id == cat["id"]:
                        class_names.append(cat["name"])
            print(f"{file_name}: {class_names}")
```
